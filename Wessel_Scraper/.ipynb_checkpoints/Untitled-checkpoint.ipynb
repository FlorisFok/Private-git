```python
import urllib.request
import urllib.parse
import re

site = 'applikon-biotechnology.com'
def scrape(url):
    pars = None
    try:
        url1 = 'http://www.' + url
        values = {'s':'basics',
                  'submit':'search'}
        data = urllib.parse.urlencode(values)
        data = data.encode('utf-8')
        req = urllib.request.Request(url1, data)
        resp = urllib.request.urlopen(req)
        respData = resp.read()
#         pars = re.findall(r"<p>(.*?)</p>", str(respData))

    except:
        try:
            url2 = 'http://' + url
            req = urllib.request.Request(url2, headers={'User-Agent' : "Magic Browser"}) 
            con = urllib.request.urlopen( req )
            respData = con.read()
#             pars = re.findall(r"<p>(.*?)</p>", str(respData))
        except:
            print('error:  '+url)
    return respData
```

```python
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
class Scraper():
    """docstring"""
    def __init__(self, my_url):
        self.my_url = my_url
        uClient = uReq(self.my_url)
        page_html = uClient.read()
        uClient.close()
        self.page_soup = soup(page_html,"html.parser")


    def fish(self,htlm_class,item):
        containers = self.page_soup.findAll(htlm_class, item)
        return containers




if __name__ == '__main__':
    my_url = "https://toto.nederlandseloterij.nl/nl"
    page = Scraper(my_url)
    htlm_class = "div"
    item = {"class":"common eventRow"}
    containers = page.fish(htlm_class,item)
    container = containers[1]
    try:
        odd_container = container.findALl('section', {'class':'infoSection details'})
        value = odd_container[0].text.strip()
    except:
        value = 'Not avalible'

    print(value)
    print('Name:')
    print(container.div.div.a.img["title"])
```
